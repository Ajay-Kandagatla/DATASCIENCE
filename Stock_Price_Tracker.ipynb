Stock Price Tracker


ðŸŽ¯ Objective
Track the historical price of a selected stock (e.g., Apple â€“ AAPL,'TSLA', 'MSFT', 'INFY.NS'), analyze trends, and visualize price movement and moving averages using Python.

```python
# Importing Required Libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Define the stock and date range
stock_symbol = 'AAPL'  # You can change to 'TSLA', 'MSFT', 'INFY.NS', etc.
start_date = '2023-01-01'
end_date = '2024-12-31'
```

```python
# Step 2: Fetch stock data
stock_data = yf.download(stock_symbol, start=start_date, end=end_date)
```

```python
# Step 3: Show basic info
print("Head of the Data:")
print(stock_data.head(10))
```

```python
# Step 4: Plot Closing Price
plt.figure(figsize=(12, 6))
plt.plot(stock_data['Close'], label='Closing Price')
plt.title(f'{stock_symbol} Closing Price')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 5: Calculate Moving Average (Optional)
stock_data['MA50'] = stock_data['Close'].rolling(window=50).mean()
```

```python
# Step 6: Plot Closing Price and MA50
plt.figure(figsize=(12, 6))
plt.plot(stock_data['Close'], label='Close Price')
plt.plot(stock_data['MA50'], label='50-Day MA', color='orange')
plt.title(f'{stock_symbol} Price & 50-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```
